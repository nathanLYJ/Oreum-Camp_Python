#### CSS 양식
```html
<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>css 작성 위치</title>
    <!-- 내부 CSS -->
    <style>
        p {
            color: red;
        }
    </style>
    <!-- 외부 CSS -->
    <link rel="stylesheet" href="002.css">
</head>
<body>  
    <header>
        <h1>hello</h1>
         <p>hello</p>
    <!-- 인라인 CSS -->
    <p style="color: green;">hello</p>
</body>
    </header>
    <main>
        <section>
            <h2>hello</h2>
            <p>hello</p>
        </section>
        <section>
            <h2>hello</h2>
            <p>hello</p>
        </section>
    </main>
    <footer>
        <h2>hello</h2>
        <p>hello</p>
    </footer>
</body>
</html>
```
* `전체 내용 실행한후 마지막 CSS 적용`
* 태그 속성 작성 할때 ^ 붙이면 뒤에 속성 아닌것 을 의미
    - class^="button"

* id 선택자
    - '#' 으로 시작
    - 문서 내에서 유일한 요소를 선택해야함
        - 문서내에서 a 태그로 호출할 때 위치 구분하기위해서
    - 선택자 이름은 대소문자 구분
* class 선택자
    - '.' 으로 시작
    - 내용 꾸미기 할 때 사용
    - 종복으로 사용가능
    - 여러개의 요소에 동시에 적용할 수 있습니다
