````python
객체지향 프로그래밍 구조 설명 (게임 캐릭터 예시)

클래스 (Class):
클래스는 객체의 설계도입니다. 게임 캐릭터를 만들기 위한 기본 틀이라고 생각하세요.
객체 (Object):
클래스를 바탕으로 실제 만들어진 캐릭터입니다. 예를 들어, "히어로"라는 이름의 전사 캐릭터가 객체입니다.
속성 (Attributes):
캐릭터의 특성을 나타냅니다. 이름, 레벨, 체력 등이 속성에 해당합니다.
메서드 (Methods):
캐릭터가 할 수 있는 행동들입니다. 공격하기, 레벨업 등이 메서드에 해당합니다.
상속 (Inheritance):
기본 캐릭터의 특성을 이어받아 특정 직업의 캐릭터를 만드는 것입니다.
다형성 (Polymorphism):
같은 이름의 메서드가 객체에 따라 다르게 동작하는 것입니다. 예를 들어, 전사와 마법사의 '공격' 메서드는 다르게 동작합니다.

도식화:

[클래스 다이어그램]

+-------------------+
|      캐릭터       |
+-------------------+
| - 이름: 문자열     |
| - 레벨: 정수      |
| - 체력: 정수      |
+-------------------+
| + 레벨업()        |
| + 상태보기()      |
+-------------------+
         ^
         |
  +------+------+
  |             |
+--------+  +--------+
|  전사   |  | 마법사  |
+--------+  +--------+
| - 공격력 |  | - 마력 |
+--------+  +--------+
| + 공격() |  | + 공격()|
+--------+  +--------+

[객체 생성 예시]

캐릭터 (추상적)
    |
    +-- 전사: 히어로 (구체적 객체)
    |
    +-- 마법사: 메이지 (구체적 객체)
상세 설명:

캐릭터 클래스 (기본 클래스)

모든 캐릭터가 공통으로 가지는 속성과 메서드를 정의합니다.
속성: 이름, 레벨, 체력
메서드: 레벨업(), 상태보기()


전사 클래스 (캐릭터 클래스를 상속)

캐릭터의 모든 특성을 상속받고, 전사만의 특성을 추가합니다.
추가 속성: 공격력
오버라이딩 메서드: 공격() - 전사 스타일의 공격


마법사 클래스 (캐릭터 클래스를 상속)

캐릭터의 모든 특성을 상속받고, 마법사만의 특성을 추가합니다.
추가 속성: 마력
오버라이딩 메서드: 공격() - 마법사 스타일의 공격


객체 생성

전사 클래스로부터 "히어로"라는 이름의 객체를 생성할 수 있습니다.
마법사 클래스로부터 "메이지"라는 이름의 객체를 생성할 수 있습니다.
````

````python
from abc import ABC, abstractmethod

class Character(ABC):
    def __init__(self, name, level=1, health=100):
        self.name = name
        self.level = level
        self.health = health

    def level_up(self):
        self.level += 1
        print(f"{self.name}의 레벨이 {self.level}로 올랐습니다!")

    def show_status(self):
        print(f"이름: {self.name}, 레벨: {self.level}, 체력: {self.health}")

    @abstractmethod
    def attack(self):
        pass

class Warrior(Character):
    def __init__(self, name, level=1, health=100, attack_power=10):
        super().__init__(name, level, health)
        self.attack_power = attack_power

    def attack(self):
        print(f"{self.name} 전사가 {self.attack_power}의 힘으로 검을 휘둘렀습니다!")

class Mage(Character):
    def __init__(self, name, level=1, health=80, mana=50):
        super().__init__(name, level, health)
        self.mana = mana

    def attack(self):
        mana_used = 10
        self.mana -= mana_used
        print(f"{self.name} 마법사가 {mana_used}의 마나를 사용하여 화염구를 던졌습니다!")

# 객체 생성 및 사용 예시
hero = Warrior("히어로")
mage = Mage("메이지")

# 전사 캐릭터 사용
hero.show_status()
hero.attack()
hero.level_up()
hero.show_status()

print("\n")

# 마법사 캐릭터 사용
mage.show_status()
mage.attack()
mage.level_up()
mage.show_status()
````

````python
class Point:


    def __init__(self, x, y):
        self.x = x
        self.y = y


    def __add__(self, other):
        if isinstance(other, Point):
            return Point((self.x + other.x), (self.y + other.y))
        elif isinstance(other, int) or isinstance(other, float):
            return Point((self.x + other), (self.y + other))


    def distance(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        return (dx**2 + dy**2)**0.5


dot1 = Point(10, 20)
dot2 = Point(20, 30)


dot3 = dot1 + 10
print(dot1.x, dot1.y) # 출력: 10 20
print(dot3.x, dot3.y) # 출력: 20 30


dot4 = dot1 + dot2
print(dot4.x, dot4.y) # 출력: 30 50
print(dot4.distance(dot1)) # 출력: 36.05551275463989
````
